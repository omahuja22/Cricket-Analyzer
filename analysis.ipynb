```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data.csv")

print("📊 Columns in dataset:\n", df.columns.tolist())

# -------- AUTO DETECT COLUMNS --------
batter_col = None
runs_col = None
bowler_col = None
over_col = None

for col in df.columns:
    col_lower = col.lower()
    
    if col_lower in ['batter', 'batsman', 'striker', 'player']:
        batter_col = col
        
    if col_lower in ['runs_off_bat', 'batsman_runs', 'runs', 'total_runs']:
        runs_col = col
        
    if col_lower == 'bowler':
        bowler_col = col
        
    if col_lower == 'over':
        over_col = col

print("\n✅ Detected Columns:")
print("Batter:", batter_col)
print("Runs:", runs_col)
print("Bowler:", bowler_col)
print("Over:", over_col)

if None in [batter_col, runs_col, bowler_col, over_col]:
    print("\n❌ ERROR: Column detection failed.")
    print("👉 Please check column names above and update manually.")
else:
    batsman_runs = df.groupby(batter_col)[runs_col].sum().sort_values(ascending=False).head(5)
    
    balls = df.groupby(batter_col).size()
    strike_rate = (batsman_runs / balls) * 100

    bowler_runs = df.groupby(bowler_col)[runs_col].sum()
    bowler_balls = df.groupby(bowler_col).size()
    economy = (bowler_runs / bowler_balls) * 6

    runs_per_over = df.groupby(over_col)[runs_col].sum()

    plt.figure()
    batsman_runs.plot(kind='bar', title="🏏 Top 5 Batsmen")
    plt.xlabel("Player")
    plt.ylabel("Runs")
    plt.show()

    plt.figure()
    runs_per_over.plot(kind='line', marker='o', title="📈 Run Rate per Over")
    plt.xlabel("Over")
    plt.ylabel("Runs")
    plt.show()

    print("\n🔍 INSIGHTS:")
    print("Top Batsman:", batsman_runs.idxmax())
    print("Best Bowler (Lowest Economy):", economy.idxmin())
    print("Highest Strike Rate Player:", strike_rate.idxmax())

    print("\n📌 Observations:")
    print("- Top batsman contributed significantly to total score")
    print("- Economical bowler controlled the run flow")
    print("- Run rate variation shows match momentum shifts")
```
